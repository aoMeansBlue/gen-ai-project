```python
import matplotlib.pyplot as plt
import numpy as np

def draw_charmander_lineart():
    fig, ax = plt.subplots(figsize=(6, 6))
    
    # Body outline (ellipse)
    body_x = np.linspace(-1, 1, 100)
    body_y = 1.2 * np.sqrt(1 - (body_x**2))  # upper body half
    ax.plot(body_x, body_y - 0.5, 'k', lw=2)
    ax.plot(body_x, -body_y - 0.5, 'k', lw=2)

    # Head (smaller circle)
    head_theta = np.linspace(0, 2*np.pi, 100)
    head_x = 0.5 * np.cos(head_theta)
    head_y = 0.5 * np.sin(head_theta)
    ax.plot(head_x, head_y + 1.0, 'k', lw=2)

    # Eye
    ax.plot([0.2, 0.2], [1.1, 1.2], 'k', lw=2)

    # Arm
    ax.plot([-0.6, -0.9], [0.0, 0.3], 'k', lw=2)
    ax.plot([0.6, 0.9], [0.0, 0.3], 'k', lw=2)

    # Legs
    ax.plot([-0.5, -0.7], [-1.2, -1.0], 'k', lw=2)
    ax.plot([0.5, 0.7], [-1.2, -1.0], 'k', lw=2)

    # Tail
    tail_x = np.linspace(1.0, 2.0, 50)
    tail_y = 0.3 * np.sin(2 * np.pi * tail_x) - 0.5
    ax.plot(tail_x, tail_y, 'k', lw=2)

    # Flame tip
    flame_x = [2.0, 2.2, 2.1, 2.3, 2.0]
    flame_y = [-0.5, -0.3, -0.7, -0.4, -0.5]
    ax.plot(flame_x, flame_y, 'k', lw=2)

    # Finishing touches
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title("Simple Line-Art Charmander", fontsize=14)
    plt.show()

# Example usage:
draw_charmander_lineart()
```
